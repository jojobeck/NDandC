## eqation derived from Vialov-profile

defined by  $$F(c)=h^m-\gamma \tau h-a_0 $$
    with $m=2(n+1)=8$ for typical n=3 (Glens flow law). with $\dot{F(c)=0}$ This gives a critical minimal surface height of
$$ h_c = (\frac{\gamma\tau}{m})^{\frac{1}{m-1}}$$. 
With F(c) =0 gives the critical surface mass balance 
$$a _{0c}= \frac{\gamma \tau}{m}^{(\frac{m}{m-1})}(1-m)$$

```python
gammaT=5
m = 8
aoc=(gammaT/m)**(m/(m-1))*(1-m)
h_c = (gammaT/m)**(1/(m-1))
def x_dot(r,x):
    return(x**8-gammaT*x-r)
n = 1000
x = np.linspace(0,1.5,n)
r_s=np.array([aoc-1,aoc,aoc+2])
```

# R-tipping

defined by  $$dxdt=(x+\lambda)^2 -\mu $$ and $$\lambda = r\cdot t$$ with $\mu>0$


```python
def x_dot(mu,l,x):
    return((x+l)**2-mu)
```

phase space fot constant $\lambda$ and $\mu = 3$

```python

n = 1000
x = np.linspace(-16,16,n)
l_s=np.array([0,3,6])
dt = np.linspace(0,6,n)
mu =3
f,ax = plt .subplots(1,3, figsize=(12,8))
for i,l in enumerate(l_s):
    ll =l*dt

    ax[i].plot(x,x_dot(mu,ll,x),label = 'r = '+str(l))
    ax[i].plot(x,x_dot(mu,l,x),'r--',label = '$r_{konst}$ = '+str(l))
    
    ax[i].plot(x,x*0,'k--',alpha =0.5)    
    ax[i].legend()
```


First find the equilibrium points by letting  dxdt=0 .
This gives  $$x_eq=\pm \sqrt{\mu} -\lambda $$
$$x_eq=\pm \sqrt{\mu} -rt $$


or finding the critical $\lambda$ for $dx/dt=0$
$$ \lambda = \pm\sqrt\mu-x $$

$$\frac{dF}{dx}=2(x+\lambda) $$
and looking at it at the equilibrium point
$$\frac{dF}{dx}|_{x=x_eq}=\pm 2\sqrt{\mu}=0$$
thus $\mu = 0$ when bifutcation occurs.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(z,t,r):
    x0, l0 = z[0], z[1]
    mu = 3
    dxdt = (x0 +l0)**2 -mu
    dldt = r    
    dzdt = [dxdt,dldt]
    return dzdt
mu = 3
   

# initial condition
v=[-10,2]
n=100000
# time points
t = np.linspace(0,5,n)
r = 1.5
# solve ODE
y = odeint(model,v,t,args=(r,))



x = np.zeros(len(y))
l= np.zeros(len(y))

for i in range(len(y)):
    x[i]=y[i][0]
    l[i]=y[i][1]   
    
A = -1*(mu-r)**0.5-x
B = (mu-r)**0.5-x

    
f,ax = plt.subplots(1,1)
ax.plot(x,l)
ax.plot(x,mu**0.5-x, 'k--',alpha =0.5)
ax.plot(x,A, 'r--',alpha =0.5,label ='A invariant line')
ax.plot(x,B, 'r--',alpha =0.5,label ='B invariant line')


ax.plot(x,-1*mu**0.5-x, 'k--',alpha =0.8)

ax.set_ylabel('$\lambda$')
ax.set_xlabel('x')
ax.legend()
```
